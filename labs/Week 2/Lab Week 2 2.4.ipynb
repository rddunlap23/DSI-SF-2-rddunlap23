```python
#Import all needed libraries for lab
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
#Load File to dataframe
file_path = '/Users/ryandunlap/Desktop/DSI-SF-2/datasets/rock_songs/rock.csv'
df = pd.read_csv(file_path)
```

```python
#Start to explore the data checking first to understand data types and completeness
df.info()
df.head(5)
```

```python
#Rename Columns
df.rename(columns={'Song Clean':'Song_Title',
                   'ARTIST CLEAN':'Artist',
                   'Combined':'Song_Title_and_Artist',
                   'Release Year':'Year_Song_Released',
                   'PlayCount':'Times_Played',
                   'F*G':'First_*_Times_Played'
    }, inplace=True)


df.head()
```

```python
#Created function to print measures of central tendency and dispersion. 
def describe_data(df,column):
    mean = np.mean(df[column])
    median = np.median(df[column])
    std = np.std(df[column])
    mode = stats.mode(df[column])[0][0]
    var = np.var(df[column])
    rng = np.ptp(df[column])
    
    print 'Column: %s' %(column)
    print 'Mean: %s, Median: %s, Mode :%s' %(mean,median,mode)
    print 'Standard Deviation :%s,  Variance: %s, Range: %s\n' %(std,var,rng)
  
describe_data(df,'First?')
describe_data(df,'Year?')
describe_data(df,'Times_Played')
describe_data(df,'First_*_Times_Played')
```

```python
#Creating function to append normalized versions of column
def norm_column(df,column):
    n_name = column + '_Norm'
    df[n_name] = (df[column] -  df[column].mean()) / df[column].std()
    return df

df = norm_column(df,'Year?')
df = norm_column(df,'Times_Played')
df = norm_column(df,'First_*_Times_Played')

df.head()
```

```python
#1 : Top 20 songs by playcount
df.sort_values('Times_Played',ascending=False).head(20)
```

```python
# Groups by year and sums on times played to see songs played per year
year_count = df[['Year_Song_Released','Times_Played']].groupby(['Year_Song_Released']).agg(['sum', 'count'])
year_count.sort_values([(u'Times_Played',u'sum')],ascending=False).head(5)
```
